# Splitting data into with binary scenario

```python
import os, shutil
import numpy as np
```

```python
path_ori_benign = 'original/benign/SOB'
path_ori_malignant = 'original/malignant/SOB'
path_subclass_scenario  = 'subclass_scenario'
```

```python
list_magnification = ['40X','100X','200X','400X']
sub_class_benign = ['adenosis', 'fibroadenoma' , 'phyllodes_tumor', 'tubular_adenoma']
sub_class_malignant = ['ductal_carcinoma', 'lobular_carcinoma' , 'mucinous_carcinoma', 'papillary_carcinoma']
binary_class = ['benign', 'malignant']
subclasses = ['tubular_adenoma', 'phyllodes_tumor', 'papillary_carcinoma',
             'mucinous_carcinoma', 'lobular_carcinoma', 'fibroadenoma',
             'ductal_carcinoma', 'adenosis']
path_subclass_scenario = 'subclass_scenario'
```

get list dir from every sub class

```python
list_dir_ori_adenosis = [x for x in os.listdir(os.path.join(path_ori_benign, sub_class_benign[0]))]
list_dir_ori_fibroadenoma = [x for x in os.listdir(os.path.join(path_ori_benign, sub_class_benign[1]))]
list_dir_ori_phyllodes_tumor = [x for x in os.listdir(os.path.join(path_ori_benign, sub_class_benign[2]))]
list_dir_ori_tubular_adenoma = [x for x in os.listdir(os.path.join(path_ori_benign, sub_class_benign[3]))]
```

```python
list_dir_ori_ductal_carcinoma = [x for x in os.listdir(os.path.join(path_ori_malignant, sub_class_malignant[0]))]
list_dir_ori_lobular_carcinoma = [x for x in os.listdir(os.path.join(path_ori_malignant, sub_class_malignant[1]))]
list_dir_ori_mucinous_carcinoma = [x for x in os.listdir(os.path.join(path_ori_malignant, sub_class_malignant[2]))]
list_dir_ori_papillary_carcinoma = [x for x in os.listdir(os.path.join(path_ori_malignant, sub_class_malignant[3]))]
```

function to get list file in every sub class

```python
def file_splitting(list_dir_subclass, path_ori_class, list_magnification, dir_name_subclass):
    list_file_ori_dir = []
    for subclass_dir in list_dir_subclass:
        for x in list_magnification:
            dir_path = os.path.join(path_ori_class, dir_name_subclass, subclass_dir, x)
            for (dirpath, dirnames, filenames) in os.walk(dir_path):
                for file in filenames:
                    list_file_ori_dir.append(os.path.join(dirpath, file))
    return list_file_ori_dir
```

get list filenames on every sub class

```python
list_file_adenosis = file_splitting(list_dir_ori_adenosis, path_ori_benign, list_magnification, sub_class_benign[0])
list_file_fibroadenoma = file_splitting(list_dir_ori_fibroadenoma, path_ori_benign, list_magnification, sub_class_benign[1])
list_file_phyllodes_tumor = file_splitting(list_dir_ori_phyllodes_tumor, path_ori_benign, list_magnification, sub_class_benign[2])
list_file_tubular_adenoma = file_splitting(list_dir_ori_tubular_adenoma, path_ori_benign, list_magnification, sub_class_benign[3])
```

```python
list_file_ductal_carcinoma = file_splitting(list_dir_ori_ductal_carcinoma, path_ori_malignant, list_magnification, sub_class_malignant[0])
list_file_lobular_carcinoma = file_splitting(list_dir_ori_lobular_carcinoma, path_ori_malignant, list_magnification, sub_class_malignant[1])
list_file_mucinous_carcinoma = file_splitting(list_dir_ori_mucinous_carcinoma, path_ori_malignant, list_magnification, sub_class_malignant[2])
list_file_papillary_carcinoma = file_splitting(list_dir_ori_papillary_carcinoma, path_ori_malignant, list_magnification, sub_class_malignant[3])
```

```python
total_benign = len(list_file_adenosis) + len(list_file_fibroadenoma) + len(list_file_phyllodes_tumor) + len(list_file_tubular_adenoma)
print("adenosis : {} , fibroadenoma : {}, phyllodes_tumor : {}, tubular_adenoma : {} \ntotal_benign: {}".format(
           len(list_file_adenosis), len(list_file_fibroadenoma), len(list_file_phyllodes_tumor), len(list_file_tubular_adenoma), (
               total_benign
           ) 
        )
     )
```

```python
total_malignant = len(list_file_ductal_carcinoma) + len(list_file_lobular_carcinoma) + len(list_file_mucinous_carcinoma) + len(list_file_papillary_carcinoma)
print("ductal_carcinoma : {} , lobular_carcinoma : {}, mucinous_carcinoma : {}, papillary_carcinoma : {} \ntotal_malignant: {}".format(
           len(list_file_ductal_carcinoma), len(list_file_lobular_carcinoma), len(list_file_mucinous_carcinoma), len(list_file_papillary_carcinoma), (
               total_malignant
           ) 
        )
     )
```

10% function percentage for test dataset

```python
test_len = lambda len_x: int((10/100)*len(len_x))
```

```python
list_file_adenosis[0].split("\\")
```

magnification splitting

```python
list_subclasses = [
    list_file_ductal_carcinoma,list_file_lobular_carcinoma,
    list_file_mucinous_carcinoma,list_file_papillary_carcinoma,
    list_file_adenosis, list_file_fibroadenoma,
    list_file_phyllodes_tumor,list_file_tubular_adenoma]
dict_magnification_subclasses = {}
for sub_class in list_subclasses:
    for file in sub_class:
        lvl = file.split("\\")[3]
        scl = file.split("/")[2].split("\\")[1]
        if lvl not in dict_magnification_subclasses:
            dict_magnification_subclasses[lvl] = {}
        if scl not in dict_magnification_subclasses[lvl]:
            dict_magnification_subclasses[lvl][scl] = []
        dict_magnification_subclasses[lvl][scl].append(file)
```

copy each images from ori dataset to test dataset for each magnification 

# Copy Test Image

```python
already_test = []
for magnification in dict_magnification_subclasses:
    for sub_class in dict_magnification_subclasses[magnification]:
        choosed = dict_magnification_subclasses[magnification][sub_class]
        len_test = test_len(choosed)
        rand = np.random.choice(len(choosed), len_test, False)
        for image in rand:
            shutil.copy(choosed[image], os.path.join(path_subclass_scenario, 'test', magnification, sub_class))
            already_test.append(choosed[image])
```

```python
print("test size {}".format(len(already_test)))
```

# Copy Val Image

```python
val_len = lambda len_x: int((10/100)*len(len_x))
```

```python
dict_magnification_subclasses = {}
for sub_class in list_subclasses:
    for file in sub_class:
        lvl = file.split("\\")[3]
        scl = file.split("/")[2].split("\\")[1]
        if file in already_test:
            continue
        if lvl not in dict_magnification_subclasses:
            dict_magnification_subclasses[lvl] = {}
        if scl not in dict_magnification_subclasses[lvl]:
            dict_magnification_subclasses[lvl][scl] = []
        dict_magnification_subclasses[lvl][scl].append(file)
```

```python
already_val = []
for magnification in dict_magnification_subclasses:
    for sub_class in dict_magnification_subclasses[magnification]:
        choosed = dict_magnification_subclasses[magnification][sub_class]
        len_test = val_len(choosed)
        rand = np.random.choice(len(choosed), len_test, False)
        for image in rand:
            shutil.copy(choosed[image], os.path.join(path_subclass_scenario, 'val', magnification, sub_class))
            already_val.append(choosed[image])
```

```python
print("validation size {}".format(len(already_val)))
```

# Copy Train Image

```python
dict_magnification_subclasses = {}
for sub_class in list_subclasses:
    for file in sub_class:
        lvl = file.split("\\")[3]
        scl = file.split("/")[2].split("\\")[1]
        if file in already_test or file in already_val:
            continue
        if lvl not in dict_magnification_subclasses:
            dict_magnification_subclasses[lvl] = {}
        if scl not in dict_magnification_subclasses[lvl]:
            dict_magnification_subclasses[lvl][scl] = []
        dict_magnification_subclasses[lvl][scl].append(file)
```

```python
already_train = []
for magnification in dict_magnification_subclasses:
    for sub_class in dict_magnification_subclasses[magnification]:
        choosed = dict_magnification_subclasses[magnification][sub_class]
        for image in choosed:
            shutil.copy(image, os.path.join(path_subclass_scenario, 'train', magnification, sub_class))
            already_train.append(image)
```

```python
print("training size {}".format(len(already_train)))
```
