```
import json

with open('training_1_400.json') as f:
  training_400_main = json.load(f)

with open('training_1_400_subclass.json') as f:
  training_400_sub = json.load(f)

with open('training_1_200.json') as f:
  training_200_main = json.load(f)

with open('training_1_200_subclass.json') as f:
  training_200_sub = json.load(f)

with open('training_1_100.json') as f:
  training_100_main = json.load(f)

with open('training_1_100_subclass.json') as f:
  training_100_sub = json.load(f)

with open('training_1_40.json') as f:
  training_40_main = json.load(f)

with open('training_1_40_subclass.json') as f:
  training_40_sub = json.load(f)
```

```
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(30,7))

ax = plt.subplot(1,2,1)
sns.set(font_scale=2)
sns.lineplot(range(0, len(training_400_main['loss'])), training_400_main['loss'], linestyle='-', marker='^', label="400X")
sns.lineplot(range(0, len(training_200_main['loss'])), training_200_main['loss'], linestyle='-', marker='o', label="200X")
sns.lineplot(range(0, len(training_100_main['loss'])), training_100_main['loss'], linestyle='-', marker='>', label="100X")
sns.lineplot(range(0, len(training_40_main['loss'])), training_40_main['loss'], linestyle='-', marker='v', label="40X")
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.ylim(0, 0.02)
plt.title('Nilai loss pada Main Class')
ax.xaxis.grid()
ax = plt.subplot(1,2,2)
sns.set(font_scale=2)
sns.lineplot(range(0, len(training_400_sub['loss'])), training_400_sub['loss'], linestyle='-', marker='^', label="400X")
sns.lineplot(range(0, len(training_200_sub['loss'])), training_200_sub['loss'], linestyle='-', marker='o', label="200X")
sns.lineplot(range(0, len(training_100_sub['loss'])), training_100_sub['loss'], linestyle='-', marker='>', label="100X")
sns.lineplot(range(0, len(training_40_sub['loss'])), training_40_sub['loss'], linestyle='-', marker='v', label="40X")
plt.xlabel('Epoch')
plt.title('Nilai loss pada Sub Class')
plt.ylabel('Loss')
plt.ylim(0, 0.02)
ax.xaxis.grid()
plt.show()
```
